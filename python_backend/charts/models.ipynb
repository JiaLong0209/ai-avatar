```python
import matplotlib.pyplot as plt
import numpy as np

# --- 1. 數據準備 ---
datasets = ['HumanML3D', 'KIT-ML']
models = ['MotionDiffuse', 'T2M-GPT', 'Free-T2M']

# FID 分數數據 (越低越好)
# HumanML3D: MotionDiffuse (0.630), Free-T2M (0.060), T2M-GPT (0.141)
# KIT-ML: MotionDiffuse (1.954), Free-T2M (0.514), T2M-GPT (0.150)
fid_humanml = [0.630, 0.141, 0.060]
fid_kit = [1.954, 0.514, 0.150]

x = np.arange(len(datasets))  # 標籤位置
width = 0.25  # 長條寬度

# --- 2. 圖表初始化 (白底設計) ---
fig, ax = plt.subplots(figsize=(12, 7), facecolor='white')
ax.set_facecolor('white')

# --- 3. 繪製分組長條圖 ---
# 使用指定顏色：灰色 (MotionDiffuse), 金色 (Free-T2M), 橘色 (T2M-GPT)
colors = ['#9E9E9E', '#FFD700', '#FF8C00']

rects1 = ax.bar(x - width, [fid_humanml[0], fid_kit[0]], width, label=models[0], color=colors[0], edgecolor='#333333')
rects2 = ax.bar(x, [fid_humanml[1], fid_kit[1]], width, label=models[1], color=colors[1], edgecolor='#333333')
rects3 = ax.bar(x + width, [fid_humanml[2], fid_kit[2]], width, label=models[2], color=colors[2], edgecolor='#333333')

# --- 4. 文字與軸設定 ---
ax.set_ylabel('FID Score (Lower is Better)', fontsize=14, fontweight='bold', color='#333333')
ax.set_title('FID Performance Comparison Across Datasets', fontsize=18, fontweight='bold', pad=20, color='#333333')
ax.set_xticks(x)
ax.set_xticklabels(datasets, fontsize=14, fontweight='bold')
ax.legend(fontsize=12, frameon=True, shadow=True)

# 移除上方與右方邊框
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color('#333333')
ax.spines['bottom'].set_color('#333333')

# --- 5. 在長條上方加上數值標籤 ---
def autolabel(rects):
    """在每個 bar 上方顯示數值"""
    for rect in rects:
        height = rect.get_height()
        ax.annotate(f'{height:.3f}',
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 5),  # 3點垂直偏移
                    textcoords="offset points",
                    ha='center', va='bottom', fontsize=11, fontweight='bold')

autolabel(rects1)
autolabel(rects2)
autolabel(rects3)

# --- 6. 調整佈局與存檔 ---
plt.tight_layout()
plt.savefig('fid_comparison_v2.png', dpi=300)
print("圖表已生成：fid_comparison_v2.png")
# plt.show(
```
